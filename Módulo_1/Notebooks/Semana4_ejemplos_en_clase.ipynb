**Ejercicio 1**. Escribir un código que tome una lista de cadenas y que imprima la cadena más larga de la lista. Si fuese una tupla en lugar de una lista, ¿cambiaría algo?


```python
lista_cadenas = ['a', 'aaa', 'apapap', 'ddsd', '']

elemento_mas_largo = lista_cadenas[0]
for cadena in lista_cadenas:
  if len(cadena) > len(elemento_mas_largo):
    elemento_mas_largo = cadena

print(elemento_mas_largo)
```

```python
# No cambia si es una tupla
lista_cadenas = ('a', 'aaa', 'apapap', 'ddsd', '')

elemento_mas_largo = lista_cadenas[0]
for cadena in lista_cadenas:
  if len(cadena) > len(elemento_mas_largo):
    elemento_mas_largo = cadena

print(elemento_mas_largo)
```

**Ejercicio 2**. Escribir un código cree una lista de los números del 1 al 10  después la modifique de tal manera que los números impares sean reemplazados por el valor `None`. Si fuese una tupla en lugar de una lista, ¿cambiaría algo?


```python
numeros = []

for i in range(10):
  if ((i+1)%2==1):  
    numeros.append(None)
  if ((i+1)%2!=1):
    numeros.append(i+1)


for i, n in enumerate(numeros):
  if (n%2 == 1):
    numeros[i] = None

numeros
```

```python
numeros = []

for i in range(10):
  numeros.append(i+1)

numeros = tuple(numeros)

for i, n in enumerate(numeros):
  if (n%2 == 1):
    numeros[i] = None

numeros
```

**Ejercicio 5**. A partir de dos listas, una de nombres y otra de fechas de cumpleaños, creá un diccionario donde las claves sean los nombres y los valores los cumpleaños. Usá tu diccionario para mostrar cuándo es el cumpleaños de Juan.
\
*Extra:* hacelo usando `zip()`.

```python
nombres = ['Alan', 'Brenda', 'Tomás', 'Juan', 'Ana', 'Sol']
fechas = ['10/11', '03/05', '17/01', '08/09', '25/04', '10/11']

dic_fechas = {}
for i in range(len(nombres)):
  dic_fechas[nombres[i]] = fechas[i]

print(dic_fechas)
print(dic_fechas['Juan'])
```

```python
# usando zip()
nombres = ['Alan', 'Brenda', 'Tomás', 'Juan', 'Ana', 'Sol']
fechas = ['10/11', '03/05', '17/01', '08/09', '25/04', '10/11']

dic_fechas = dict(zip(nombres, fechas))

print(dic_fechas)
print(dic_fechas['Juan'])
```

**Ejercicio 6**. Dado el siguiente diccionario de notas de un examen:


1.   Agregá las entradas de Lucas y Paula, quienes se sacaron un 8 y un 6, respectivamente.
2.   Imprimí en pantalla los alumnos que hayan aprobado (nota >= 7) junto con su nota.



```python
dic_notas = {'Alan':4, 'Brenda':9.5, 'Tomás':6.5, 'Juan':7, 'Ana':7, 'Sol':6}

dic_notas['Lucas'] = 8
dic_notas['Paula'] = 9

for clave, valor in dic_notas.items():
  if valor >= 7:
    print(clave, valor)
```

**Ejercicio 7**. Dada las listas de antes de nombres y compleaños:

1.   Usando la función `zip()`. Creá una lista de tuplas (nombre, cumpleaños)

2.   Ordená la lista de tuplas alfabéticamente según el nombre. ¿Qué pasaría sí quisieramos ordenarla por la fecha?

3. Haciendo la operación inversa del `zip()`, recuperá las dos listas originales a partir de la lista de tuplas.

```python
nombres = ['Alan', 'Brenda', 'Tomás', 'Juan', 'Ana', 'Sol']
fechas = ['10/11', '03/05', '17/01', '08/09', '25/04', '10/11']

# 1
lista_tuplas = list( zip(nombres, fechas) )

# 2
lista_tuplas = sorted(lista_tuplas)

# 3
lista_ordenada = list(zip(*lista_tuplas))
nombres_ordenados = list(lista_ordenada[0])
fechas_ordenadas = list(lista_ordenada[1])
print(nombres_ordenados)
print(fechas_ordenadas)
```
